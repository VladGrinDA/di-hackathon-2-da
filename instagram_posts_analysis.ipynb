# Viral

```python
# Re-importing necessary libraries and loading the dataset again
import pandas as pd

# Load the dataset into the 'df' dataframe
df = pd.read_csv('/content/sample_data/instagram_posts.csv')

# List of key metrics
metrics = ['Impressions', 'Reach', 'Shares', 'Likes', 'Saves']

# Calculating the mean and standard deviation for each of the key metrics
means = df[metrics].mean()
std_devs = df[metrics].std()

# Define a threshold: posts with values greater than 2 standard deviations above the mean are considered viral
viral_threshold = means + 2 * std_devs

# Filter the viral posts based on the threshold
viral_posts = df[
    (df['Impressions'] > viral_threshold['Impressions']) |
    (df['Reach'] > viral_threshold['Reach']) |
    (df['Shares'] > viral_threshold['Shares']) |
    (df['Likes'] > viral_threshold['Likes']) |
    (df['Saves'] > viral_threshold['Saves'])
]

# Display the viral posts using Google Colab's default display
from google.colab import data_table

# Display the dataframe
data_table.DataTable(viral_posts)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define viral and non-viral posts based on impressions or reach
viral_threshold = 0.9  # top 10% threshold
impression_threshold = df['Impressions'].quantile(viral_threshold)
reach_threshold = df['Reach'].quantile(viral_threshold)

# Create a new column 'Post Type' to identify viral and non-viral posts
df['Viral'] = np.where((df['Impressions'] > impression_threshold) | (df['Reach'] > reach_threshold), 'Viral', 'Non-viral')

# Step 2: Plot a pie chart to show viral vs non-viral impressions or reach
viral_counts = df['Viral'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(viral_counts, labels=viral_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen'], startangle=90)
plt.title('Viral vs Non-Viral Posts based on Impressions or Reach')
plt.show()

# Step 3: Identify the best performing non-viral posts for impressions and reach
non_viral_posts = df[df['Viral'] == 'Non-viral']
best_non_viral_impression = non_viral_posts.loc[non_viral_posts['Impressions'].idxmax()]
best_non_viral_reach = non_viral_posts.loc[non_viral_posts['Reach'].idxmax()]

# Step 4: Find common conditions for viral posts
viral_posts = df[df['Viral'] == 'Viral']
common_post_types = viral_posts['Post type'].value_counts().head()
common_publish_times = viral_posts['Publish time'].apply(lambda x: x.split(' ')[1]).value_counts().head()

# Display the best performing non-viral posts and common conditions for viral posts
best_non_viral_impression, best_non_viral_reach, common_post_types, common_publish_times
```

```python
# Step 1: Filter viral posts
viral_posts = df[df['Viral'] == 'Viral']

# Step 2: Group viral posts by 'Duration (sec)' and calculate the average reach and impressions for each duration
# Use a list instead of a tuple to select multiple columns
viral_duration_analysis = viral_posts.groupby('Duration (sec)')[['Reach', 'Impressions']].mean().sort_values(by=['Reach', 'Impressions'], ascending=False)

# Step 3: Find the duration with the highest reach and impressions
best_viral_duration = viral_duration_analysis.head(1)

# Display the result
print("Best duration for viral posts based on average Reach and Impressions:")
print(best_viral_duration)
```

# Non Viral

```python
# Step 1: Filter non-viral posts
non_viral_posts = df[df['Viral'] == 'Non-viral']

# Step 2: Group non-viral posts by 'Duration (sec)' and calculate the average reach and impressions for each duration
# Use a list instead of tuple to select multiple columns
duration_analysis = non_viral_posts.groupby('Duration (sec)')[['Reach', 'Impressions']].mean().sort_values(by=['Reach', 'Impressions'], ascending=False)

# Step 3: Find the duration with the highest reach and impressions
best_duration = duration_analysis.head(1)

# Display the result
print("Best duration for non-viral posts based on average Reach and Impressions:")
print(best_duration)
```

The steep decline in 2024 may indicate the need to revisit content strategy, as engagement seems to have fallen off significantly.

This might indicate that while the posts were visible (due to the high reach in 2023), they didn't necessarily lead to a proportionate increase in user interaction (likes, shares, saves). This could be a signal to improve content that encourages more user participation
